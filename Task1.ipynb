```python

import numpy as np

#Initializing the values
u = 8
v = 1
w = 4

#Calculating the initial error
y = (2*u - 14)**3 + (4*v - 16)**2 + (v**2 - 16)**2 + (5*w**2/w - 15)**2

#Setting learning rate and iterations
lr = 0.001
iterations = 10000
#Performing Gradient Descent
for i in range(iterations):
    y = (2*u - 14)**3 + (4*v - 16)**2 + (v**2 - 16)**2 + (5*w**2/w - 15)**2
    print(f"Iteration : {i}  | Error = {y}")
    
    
    #Calculating partial derivatives of the error
    du = (6*(2*u - 14)**2)
    dv = (8*(4*v - 16) + 2*(v**2 - 16))
    dw = (20*w**2)/(w - 15)**3 - (10*w**3)/(w - 15)**3
    
    #Updating the values of u, v and w
    u = u - lr * du
    v = v - lr * dv
    w = w - lr * dw

#Calculating the new error
y = (2*u - 14)**3 + (4*v - 16)**2 + (v**2 - 16)**2 + (5*w**2/w - 15)**2

#Printing the values
print('Final values of u,v,w after Gradient Descent: ',u,v,w)
print('The final error after Gradient Descent: ', y)
```
